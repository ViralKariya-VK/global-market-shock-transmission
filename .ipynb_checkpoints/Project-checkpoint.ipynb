A strong academic project follows a rigorous path. You should structure your final notebook or report in these five stages:
1. **Exploration & Decomposition:** Visualizing the data to separate the signal into three components: Trend, Seasonality, and Residuals (Noise).
2. **Stationarity Checks:** Testing if the statistical properties (mean, variance) remain constant over time. This usually involves the Augmented Dickey-Fuller (ADF) test.
3. **Transformation:** Making the series stationary (using Differencing or Log Transforms) so traditional models like ARIMA can work.
4. **Model Identification:** Using ACF (Auto-Correlation Function) and PACF (Partial Auto-Correlation Function) plots to manually determine the order of your model (the $p, d, q$ parameters).
5. **Evaluation:** Comparing predictions against a test set using specific metrics (RMSE, MAPE).

```python
# Convert date column DataType to Date and then
# Setting it to index

usd_inr["Date"] = pd.to_datetime(usd_inr["Date"])
usd_inr.set_index("Date", inplace=True)
```

```python
usd_inr.head()
```

## Merge the Data

### Keep only 'close' column

```python
sp500 = sp500[["Close"]]
nifty50 = nifty50[["Close"]]
usd_inr.columns = ["USDINR"]

sp500.columns = ["SP500"]
nifty50.columns = ["NIFTY"]
```

### Align all the Data for Date starting from 

```python
data = sp500.join(nifty50, how="inner")
data = data.join(usd_inr, how="inner")
data = data.dropna()
```

```python
data.head()
data.index.min()
data.index.max()
data.isna().sum()
```

```python
data.head()
```

```python
# Here, all three column are currency form, but log values are preffered for the VAR model
```

```python
returns = np.log(data / data.shift(1))
returns = returns.dropna()
```

```python
returns.head()
```

```python
returns.describe()
```

#### Visualize the logged values

```python
returns.plot(subplots=True, figsize=(10,8))
plt.show()
```

### Stationarity Checks

#### ADF Test

```python
from statsmodels.tsa.stattools import adfuller

for col in returns.columns:
    print(col, "ADF p-value:", adfuller(returns[col])[1])
```

## VAR Model

```python
from statsmodels.tsa.api import VAR

model = VAR(returns)
lag_order = model.select_order(maxlags=10)
print(lag_order.summary())
```

```python
model_fitted = model.fit(lag_order.selected_orders["aic"])
print(model_fitted.summary())
```

### Stability Check
- Must return True

```python
model_fitted.is_stable()
```
