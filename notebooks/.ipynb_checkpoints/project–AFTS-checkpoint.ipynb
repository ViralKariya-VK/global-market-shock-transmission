### Returns

```python
def transform_to_returns(df):
    """
    Converts raw prices to log returns.
    VIX is often analyzed in raw levels because it is a percentage and naturally 
    mean-reverting, but for strict VAR modeling alongside other returns, 
    we take the log difference here as well to ensure total stationarity.
    """
    # Calculate log returns: ln(Pt / P_{t-1})
    returns_df = np.log(df / df.shift(1))
    
    # Drop the first row since shifting creates a NaN for Day 1
    returns_df = returns_df.dropna()
    
    return returns_df
```

```python
# Transform to Log Returns
returns = transform_to_returns(data)
```

```python
returns.head()
```

### Stationarity Test

```python
def run_adf_test(df, significance_level=0.05):
    """
    Runs the Augmented Dickey-Fuller test on each column.
    Null Hypothesis (H0): The series has a unit root (is non-stationary).
    Alternate Hypothesis (H1): The series has no unit root (is stationary).
    """
    results = []
    
    for col in df.columns:
        # Run the ADF test using AIC to automatically select the optimal lags
        adf_stat = adfuller(df[col], autolag='AIC')
        
        p_value = adf_stat[1]
        is_stationary = "Yes" if p_value < significance_level else "No"
        
        results.append({
            'Variable': col,
            'ADF Statistic': round(adf_stat[0], 4),
            'P-Value': round(p_value, 6),
            'Stationary (5%)': is_stationary
        })
        
    results_df = pd.DataFrame(results)
    return results_df
```

```python
# Test Raw Prices

print("ADF Test on Raw Prices")
print("-" * 30)
adf_raw = run_adf_test(data)
adf_raw.head()
```

```python
# Test Log Returns
print("\n ADF Test on Log Returns")
print("-" * 30)
adf_returns = run_adf_test(returns)
adf_returns.head()
```

## Objective 1 – Do the US and Indian markets share a long-term mathematical anchor?

### Johansen Cointegration Test

Johansen Cointegration Testing was conducted to see if a long-term mathematical anchor exists between these markets. This is often called the "Drunk and His Dog" theory: both wander randomly, but a "leash" prevents them from drifting too far apart forever.

The Johansen Conintegration suggest that, it can test strictly on I(1) variables, meaning it only works on the variables where the data becomes stationary after differencing once.

```python
def select_optimal_lags(returns_df, maxlags):
    """
    Finds the optimal lag length using BIC.
    """
    model = VAR(returns_df)
    lag_results = model.select_order(maxlags=maxlags)
    
    
    # Extract specifically the BIC lag
    optimal_lag_bic = lag_results.bic
    print(f"\nOptimal Lag Selected (BIC): {optimal_lag_bic}")
    
    return optimal_lag_bic
```

```python
def run_johansen(raw_df, optimal_lag, i1_vars):
    """
    Runs the Johansen Cointegration test strictly on the I(1) variables.
    """
    
    df_i1 = raw_df[i1_vars]
    
    johansen_out = coint_johansen(df_i1, det_order=0, k_ar_diff=optimal_lag - 1)
    
    traces = johansen_out.lr1
    cvts = johansen_out.cvt[:, 1] # 5% critical values
    
    results = []
    for i in range(len(traces)):
        results.append({
            'Series': i1_vars[i],
            'Hypothesis (r = cointegrating vectors)': f'r <= {i}',
            'Trace Statistic': round(traces[i], 4),
            'Critical Value (5%)': round(cvts[i], 4),
            'Is Cointegrated?': 'Yes' if traces[i] > cvts[i] else 'No'
        })
        
    return pd.DataFrame(results)
```

```python
# BIVARIATE TEST (Just SP500 and NIFTY)
# ---------------------------------------------------------

print("Bivariate Cointegration (SP500 & NIFTY)")

bivariate_returns = returns[['SP500', 'NIFTY']]

optimal_lag_bivariate = select_optimal_lags(bivariate_returns, maxlags=10)
                                           
model = VAR(bivariate_returns)
results = model.fit(optimal_lag_bivariate)

residuals = results.resid

for col in residuals.columns:
    plot_acf(residuals[col], lags=30)
    plt.title(f"ACF of Residuals: {col}")
    plt.show()
```

The Overfitting Trap: A decision was needed on how many past days of data the model should use to predict today’s movement.

A criterion called AIC suggested 22 days of memory. However, this creates a "brittle" model with too many variables (nearly 800 parameters), leading to the model memorizing noise rather than learning patterns.

A stricter criterion called BIC suggested 3 days. This aligns with reality: a shock hits the US, India reacts the next day, and trades settle by the third day.

Final Choice: A 3-day lag was locked in as the most efficient and realistic window for market contagion.

The Whiteness Test: After fitting the model, the "leftover" noise (residuals) was checked to see if it was perfectly random.

The Problem: The noise failed the randomness test. Attempts to fix this by adding more lags (up to 100 days) were unsuccessful.

The Insight: The issue wasn't a lack of memory, but rather Volatility Clustering. Financial markets have "mood swings"—periods of high stress followed by calm.

The Solution: The 3-day lag was maintained because it captures the main price ripple. The remaining "mood swings" are identified as ARCH effects, which are better handled by a GARCH model later rather than forcing more lags into the current model.

The core engine (VECM) is now finalized with a 3-day lag. It is mathematically proven that the US and Indian markets are tied together by a long-term anchor. The model is now ready to generate Impulse Response Functions (IRF) to map out exactly how a 1% shock in the US ripples through India over the following week.

```python
results_bivariate = run_johansen(data, optimal_lag_bivariate, i1_vars=['SP500', 'NIFTY'])
results_bivariate.head()
```

```python
# ---------------------------------------------------------
# MULTIVARIATE TEST (All four I(1) Variables)
# ---------------------------------------------------------
print("--- Multivariate Cointegration (Macro Ecosystem) ---")

# Step A: Isolate the returns for the four I(1) variables
multivariate_returns = returns[['SP500', 'NIFTY', 'USDINR', 'DXY']]

# Step B: Find the optimal lag specifically for this system
optimal_lag_multi = select_optimal_lags(multivariate_returns, maxlags=10)
                                          
model = VAR(multivariate_returns)
results = model.fit(optimal_lag_multi)

residuals = results.resid

for col in residuals.columns:
    plot_acf(residuals[col], lags=30)
    plt.title(f"ACF of Residuals: {col}")
    plt.show()
```

```python
# Step C: Run Johansen using the system's specific lag length
results_multivariate = run_johansen(data, optimal_lag_multi, i1_vars=['SP500', 'NIFTY', 'USDINR', 'DXY'])
results_multivariate.head()
```

A shift in the optimal lag from $k=3$ (Bivariate) to $k=1$ (Multivariate) is observed. This is attributed to the BIC's increased penalty for complexity in higher-dimensional systems and the inclusion of currency variables (DXY, USDINR) which reduce the need for long-term equity lags. Post-estimation ACF plots confirm that Lag 1 is sufficient to eliminate serial correlation in the multivariate residuals.

## The Conclusion for the Objective
1. The "Leash" exists in isolation (Bivariate)The bivariate test shows exactly one cointegrating vector ($r=1$). This proves that even if everything else in the world (Oil, Currencies, VIX) is ignored, the S&P 500 and NIFTY 50 share a long-term gravitational pull. They cannot drift apart forever; eventually, one will pull the other back into line.
2. The "Ecosystem" is stable (Multivariate)The multivariate test shows two cointegrating vectors ($r=2$). This indicates that adding the US Dollar (DXY) and the Rupee (USDINR) into the mix doesn't break the relationship; it adds another layer of stability. The entire system—US stocks, Indian stocks, and the currencies—acts like a self-correcting machine.
3. The Model Choice is MandatoryBecause cointegration is statistically confirmed, a standard VAR (Vector Autoregression) model is now legally "wrong." A VAR would only see the daily noise and miss the long-term anchor. The analysis must proceed with a VECM (Vector Error Correction Model) to capture how the markets "correct" themselves when they get out of sync.

```python
# 1. Isolate the I(1) variables
i1_vars = ['SP500', 'NIFTY', 'USDINR', 'DXY']
df_i1 = data[i1_vars]

# 2. Build the VECM Engine
# k_ar_diff = 1 (from multivariate BIC)
# coint_rank = 2 (from Johansen results)
# deterministic = 'co' (to match the constant in our cointegration test)
vecm_model = VECM(df_i1, 
                  k_ar_diff=1, 
                  coint_rank=2, 
                  deterministic='co')

vecm_results = vecm_model.fit()

# 3. Extract the Alpha (Speed of Adjustment)
alpha_results = pd.DataFrame(vecm_results.alpha, 
                             index=i1_vars, 
                             columns=['Vector 1 Adjustment', 'Vector 2 Adjustment'])

print("--- Speed of Adjustment (Alpha Coefficients) ---")
print(alpha_results.round(5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure your VECM has been fitted in the previous step:
# vecm_results = vecm_model.fit()

def plot_custom_irf_clean(vecm_results, impulse_var='SP500', response_var='NIFTY', days=15):
    """
    Creates a publication-quality, annotated Impulse Response plot to visualize contagion.
    """
    # 1. Set a clean aesthetic theme
    sns.set_theme(style="whitegrid")
    
    # 2. Generate IRF data (let's look at 15 days, roughly 3 business weeks)
    irf = vecm_results.irf(periods=days)
    
    # 3. Identify the indices of our variables
    var_names = vecm_results.names # ['SP500', 'NIFTY', 'USDINR', 'DXY']
    imp_idx = var_names.index(impulse_var)
    res_idx = var_names.index(response_var)
    
    # 4. Extract the response values and the standard errors for confidence intervals
    response = irf.irfs[:, res_idx, imp_idx]
    stderr = irf.stderr()[:, res_idx, imp_idx]
    
    # 5. Calculate the 95% Confidence Intervals (Roughly Response +/- 2*StdErr)
    lower_ci = response - 1.96 * stderr
    upper_ci = response + 1.96 * stderr
    t = np.arange(days + 1) # Time horizon
    
    # 6. Setup the plot
    plt.figure(figsize=(8, 4))
    
    # A. Plot the baseline zero line (Equilibrium) - bold and prominent
    plt.axhline(0, color='black', linewidth=2.0, linestyle='-')
    
    # B. Shade the confidence interval (The 'Confidence Zone')
    # Using a soft alpha makes it less harsh.
    plt.fill_between(t, lower_ci, upper_ci, color='#d9534f', alpha=0.15, 
                     label='95% Confidence Zone (Significance)')
    
    # C. Plot the main response line (The 'Contagion Ripple')
    # Using a vibrant red with markers to indicate days.
    plt.plot(t, response, color='#d9534f', linewidth=3.0, marker='o', 
             markersize=6, label=f'{response_var} Response to {impulse_var} Shock')
    
    # D. Formatting for maximum readability
    plt.title(f'Daily Economic Contagion: Immediate Ripple from US to India', 
              fontsize=18, fontweight='bold', pad=20)
    plt.ylabel(f'Change in {response_var} (Log Units)', fontsize=14, labelpad=10)
    plt.xlabel('Trading Days After US Shock', fontsize=14, labelpad=10)
    
    # Set x-ticks to show every day clearly
    plt.xticks(t)
    
    # Create a nice border legend
    plt.legend(frameon=True, fontsize=11, loc='upper right')
    
    # E. Advanced Customization: Animate the Interpretation
    # Let's add an arrow to show recovery time
    significant_days = np.where(upper_ci < 0)[0]
    if len(significant_days) > 0:
        recovery_day = significant_days[-1] + 1
        plt.annotate('', xy=(recovery_day, -0.001), xytext=(recovery_day, 0.002),
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))
        plt.text(recovery_day + 0.2, 0.0015, f'Recovery Time:\n~{recovery_day} Trading Days', 
                 fontsize=11, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.5))

    # Add annotation for Immediate Impact
    plt.annotate(f'Immediate Impact: {response[1]:.5f}', xy=(1, response[1]), xytext=(2.5, response[1]-0.001),
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='#d9534f'),
                 color='#d9534f', fontsize=10, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# --- Execution ---
plot_custom_irf_clean(vecm_results, impulse_var='SP500', response_var='NIFTY', days=15)
```

This graph shows that a shock in the US market causes an immediate and powerful jump in the Indian market on Day 1. Most importantly, because the response stays positive and doesn't return to zero, it proves our Cointegration theory: the two markets are mathematically linked. When the US grows, India is pulled along for the ride and stays at that new higher level, rather than falling back.
